```python
# Test Preprocessing

from preprocess import full_preprocessing

df, basket_df, customer_df = full_preprocessing("Dataset/Retail_Transactions_Dataset.csv")

print("=== CLEANED DATAFRAME ===")
display(df.head())

print("=== BASKET DATAFRAME (One-Hot) ===")
display(basket_df.head())

print("=== CUSTOMER AGGREGATES ===")
display(customer_df.head())

print("Rows:", len(df))
print("Basket rows:", len(basket_df))
```

```python
# Test Market Basket Engine

from market_basket import (
    generate_frequent_itemsets,
    generate_rules,
    get_rules_for_product,
    generate_context_rules
)

# Step 1: Frequent Itemsets
itemsets = generate_frequent_itemsets(basket_df, min_support=0.01)
print("=== FREQUENT ITEMSETS ===")
display(itemsets.head())

# Step 2: Rules
rules = generate_rules(itemsets, metric="lift", min_threshold=1.0)
print("=== ASSOCIATION RULES ===")
display(rules.head())

# Step 3: Product-specific rules
print("=== RULES FOR PRODUCT: 'Milk' ===")
milk_rules = get_rules_for_product(rules, "Milk")
display(milk_rules.head())

# Step 4: Context-Aware Rules Example
winter_rules = generate_context_rules(df, {"Season": "Winter"}, min_support=0.01)
print("=== WINTER CONTEXT RULES ===")
display(winter_rules.head())
```
